```python
# prompt: apply cross validation now

# Assuming 'X' and 'y' are already defined from your preceding code
# For example, if 'Exited' is your target variable:
x = df.drop('Exited', axis=1)
y = df['Exited']

# Initialize a linear regression model
model = LinearRegression()

# Apply cross-validation
# cv=5 means 5-fold cross-validation
scores = cross_val_score(model, x, y, cv=5)

# The scores are negative MSE, so we take the absolute value and then the square root
rmse_scores = np.sqrt(-scores)

print("Cross-validation RMSE scores:", rmse_scores)
print("Average RMSE:", rmse_scores.mean())
print("Standard Deviation of RMSE:", rmse_scores.std())
```

```python
a = df.drop('Exited', axis=1)
b = df['Exited']
```

```python
from imblearn.over_sampling import RandomOverSampler
rs=RandomOverSampler()
x_res, y_res = rs.fit_resample(a,b)
```

```python
#train_test_split

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_res,y_res,test_size=0.2,random_state=30)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(max_depth=9)
rfc.fit(x_train, y_train)
```

```python
y_pred_train = rfc.predict(x_train)
y_pred_test  = rfc.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_test)
```

```python
accuracy_score(y_train,y_pred_train)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(rfc,x_train,y_train,cv=10).mean()
```

```python
print(classification_report(y_test,y_pred_test))
```
